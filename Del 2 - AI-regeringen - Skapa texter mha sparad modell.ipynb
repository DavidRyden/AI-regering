```python
import random
import sys
import re
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
## Ladda tidigare skapad modell

from keras.models import load_model
model = load_model('regf_master.h5')
```

```python
def sample(preds, temperature=1.0):
    preds = np.asarray(preds).astype('float64')
    preds[preds < .02] = 0.0000
    preds[12]=preds[12]*2 # punkt
    preds[65]=preds[65]*1.5 #newline
    preds = np.log(preds) / temperature
    exp_preds = np.exp(preds)
    preds = exp_preds / np.sum(exp_preds)
    probas = np.random.multinomial(1, preds, 1)
    return np.argmax(probas)
```

```python
collected_ai_texts = []  

for j in range(0,20):    
    maxlen = 200
    
    # Select a text seed at random
    #Eders Majestäter, Eders Kungliga Högheter, Herr talman, Ärade ledamöter av Sveriges riksdag
    
    add_on_text_cap = """Eders Majestäter, <br/>Eders Kungliga Högheter, <br/>Herr talman, <br/>Ärade ledamöter av Sveriges riksdag!
    <br/>Svenska folket har röstat för en ny politisk inriktning. Vi föreslår tillsättandet av en regering som enbart består av en AI-bot bestämmer att Sverige ska """

    
    add_on_text = """eders majestäter, eders kungliga högheter, herr talman, ärade ledamöter av sveriges riksdag!
    \n\nsvenska folket har röstat för en ny politisk inriktning. vi föreslår tillsättandet av en regering som enbart bestårs av en ai-bot bestämmer att sverige """
    #vi föreslår tillsättandet av en regering enbart beståendes av en ai-bot som skriver texter till i regeringens sakfrågor och som kan"
    
    
    generated_text =  add_on_text[:200]  
    ai_text=generated_text
    
    #print('--- Generating with seed: \n"' + generated_text + '"')
    
    for temperature in [0.35]:
        #print('------ temperature:', temperature)
        #sys.stdout.write(generated_text)
    
        # We generate 400 characters
        for i in range(20000):
            sampled = np.zeros((1, maxlen, len(chars)))
            for t, char in enumerate(generated_text):
                sampled[0, t, char_indices[char]] = 1.
    
            preds = model.predict(sampled, verbose=0)[0]
            next_index = sample(preds, temperature)
            next_char = chars[next_index]
    
            generated_text += next_char
            generated_text = generated_text[1:]
    
            ai_text += next_char
            
            #sys.stdout.write(next_char)
            #sys.stdout.flush()
        #print()
        
        #ai_text = '. '.join(map(lambda s: s.capitalize(), ai_text.split('. ')))
        #ai_text = '! '.join(map(lambda s: s.strip().capitalize(), ai_text.split('! ')))
        #ai_text = '? '.join(map(lambda s: s.strip().capitalize(), ai_text.split('? ')))
        #ai_text = '\n'.join(map(lambda s: s.strip().capitalize(), ai_text.split('\n')))
        collected_ai_texts += [add_on_text_cap + ai_text[200:]]
                              
```

```python
##Stor bokstav efter punkt och ny rad

for j in range(0, len(collected_ai_texts) ):
    
    result= ''
    textlist = collected_ai_texts[j].split('. ')
    
    for item in textlist:
        
        if item==textlist[0]:
            result += re.sub('([a-zA-Z])', lambda x: x.groups()[0].upper(), item, 1)
        else:
            result += '. ' + re.sub('([a-zA-Z])', lambda x: x.groups()[0].upper(), item, 1)
    
    result1 = ''
    textlist1 = result.split('\n')    
    for item in textlist1:
        if item==textlist1[0]:
            result1 += re.sub('([a-zA-Z])', lambda x: x.groups()[0].upper(), item, 1)
        else:
            result1 += '\n' + re.sub('([a-zA-Z])', lambda x: x.groups()[0].upper(), item, 1)
    #print (result1)
    collected_ai_texts[j]=result1
    
    
    
    #'. '.join(map(lambda s: s.strip().capitalize(), collected_ai_texts[j].split('. ')))
    #print(collected_ai_texts[j])
```

```python

## Sverige med stort S och lite annat fix

for j in range(0, len(collected_ai_texts) ):
    #collected_ai_texts[j] = '. '.join(map(lambda s: s.strip().capitalize(), collected_ai_texts[j].split('. ')))
    collected_ai_texts[j] = collected_ai_texts[j].replace('sverige', 'Sverige').replace('\ns', '\nS').replace('\n','<br/> <br/>')
    #print(collected_ai_texts[j])
```
